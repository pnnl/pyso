```python
df = pd.read_csv(csv_file,low_memory=False)
```

```python
print(df)
```

Function to help divide up different categories by state

```python
def print_unique_values_with_counts(df, column_name):
    # Get value counts for the specified column
    value_counts = df[column_name].value_counts()
    
    # Print each unique value and its count
    for value, count in value_counts.items():
        print(f"{value}: {count}")
```

## Filtering to only WECC states that Shat considered: WA, OR, CA, ID, MT

### 10,918 communities with 133 metrics

```python
WECC_states = ['Washington','Oregon','California','Idaho','Montana','Wyoming','Utah','Colorado','New Mexico','Nevada','Arizona']
WECC_df = df[df['State/Territory'].isin(WECC_states)]
print(WECC_df)
```

```python
print_unique_values_with_counts(WECC_df,'State/Territory')
```

## Filtering WECC communities to those that are DAC due to energy burden

### 174 communities with 133 metrics

```python
energyBurden_df = WECC_df[WECC_df['Greater than or equal to the 90th percentile for energy burden and is low income?'] == True]
```

```python
print(energyBurden_df)
```

```python
print_unique_values_with_counts(energyBurden_df,'State/Territory')
```

```python
print_unique_values_with_counts(energyBurden_df,'County Name')
```
